```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')

# 设置中文显示
plt.rcParams['axes.unicode_minus'] = False

# 创建结果文件夹
import os
PROBLEM_NUM = 3
os.makedirs(f'results/问题{PROBLEM_NUM}', exist_ok=True)

def save_result(data, description, file_format='csv'):
    """保存重要结果到文件"""
    if file_format == 'csv':
        filename = f"results/问题{PROBLEM_NUM}/{description}.csv"
        data.to_csv(filename, encoding='utf-8-sig', index=False)
    elif file_format == 'excel':
        filename = f"results/问题{PROBLEM_NUM}/{description}.xlsx"
        data.to_excel(filename, index=False, engine='openpyxl')
    print(f"已保存: {filename}")
    return filename

def save_plot(description, dpi=300, format='png'):
    """保存图表到文件"""
    filename = f"results/问题{PROBLEM_NUM}/{description}.{format}"
    plt.savefig(filename, dpi=dpi, bbox_inches='tight')
    print(f"已保存: {filename}")
    return filename

def load_data(filename, encoding='utf-8-sig'):
    """标准化数据读取"""
    filepath = f"data/{filename}"
    if filename.endswith('.csv'):
        return pd.read_csv(filepath, encoding=encoding)
    elif filename.endswith('.xlsx'):
        return pd.read_excel(filepath, engine='openpyxl')
    else:
        raise ValueError(f"不支持的文件格式: {filename}")

# 1. 数据准备和回归模型重建
def prepare_optimization_data():
    """准备优化分析数据"""
    raw_data = load_data("附件1.csv")
    data = raw_data.copy()
    
    # 基础特征工程
    data['C4烯烃收率'] = data['乙醇转化率(%)'] * data['C4烯烃选择性(%)'] / 100
    data['含HAP'] = data['HAP_mass'] > 0
    data['Co_SiO2_HAP_ratio_corrected'] = np.where(
        data['HAP_mass'] > 0,
        data['Co_SiO2_mass'] / data['HAP_mass'],
        -1
    )
    
    # 筛选含HAP样本用于建模
    analysis_data = data[data['含HAP']].copy()
    
    print(f"数据准备完成: {len(analysis_data)}个含HAP样本")
    return analysis_data

def rebuild_regression_models(data):
    """重建问题2的回归模型"""
    
    # 定义变量
    predictor_vars = ['温度', 'Co_loading', 'ethanol_conc', '总催化剂质量', 'Co_SiO2_HAP_ratio_corrected']
    target_vars = ['乙醇转化率(%)', 'C4烯烃选择性(%)', 'C4烯烃收率']
    
    models = {}
    
    for target in target_vars:
        X = data[predictor_vars].select_dtypes(include=[np.number])
        y = data[target]
        
        # 处理缺失值
        valid_mask = ~(X.isnull().any(axis=1) | y.isnull())
        X_clean = X[valid_mask]
        y_clean = y[valid_mask]
        
        if len(X_clean) < 10:
            continue
            
        # 数据标准化
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X_clean)
        
        # 建立回归模型
        model = LinearRegression()
        model.fit(X_scaled, y_clean)
        
        # 模型评估
        y_pred = model.predict(X_scaled)
        r2 = r2_score(y_clean, y_pred)
        
        models[target] = {
            'model': model,
            'scaler': scaler,
            'feature_names': X_clean.columns.tolist(),
            'r2': r2,
            'coefficients': model.coef_,
            'intercept': model.intercept_
        }
        
        print(f"{target}模型重建完成, R² = {r2:.4f}")
    
    return models

def predict_yield(models, temp, co_loading, ethanol_conc, total_catalyst, ratio):
    """预测C4烯烃收率"""
    
    # 构建输入特征
    features = np.array([[temp, co_loading, ethanol_conc, total_catalyst, ratio]])
    
    # 预测转化率和选择性
    conversion_model = models['乙醇转化率(%)']
    selectivity_model = models['C4烯烃选择性(%)']
    
    # 标准化输入
    conv_features_scaled = conversion_model['scaler'].transform(features)
    sel_features_scaled = selectivity_model['scaler'].transform(features)
    
    # 预测
    conversion_pred = conversion_model['model'].predict(conv_features_scaled)[0]
    selectivity_pred = selectivity_model['model'].predict(sel_features_scaled)[0]
    
    # 计算收率
    yield_pred = conversion_pred * selectivity_pred / 100
    
    return {
        'conversion': max(0, min(100, conversion_pred)),
        'selectivity': max(0, min(100, selectivity_pred)), 
        'yield': max(0, yield_pred)
    }

# 2. 数据驱动的最优组合识别
def find_optimal_combinations(data):
    """从现有数据中找到最优组合"""
    
    print("数据驱动优化分析:")
    print("="*50)
    
    # 按温度约束分组分析
    all_data = data.copy()
    constrained_data = data[data['温度'] < 350].copy()
    
    results = {}
    
    # 无约束优化
    max_yield_idx = all_data['C4烯烃收率'].idxmax()
    optimal_all = all_data.loc[max_yield_idx]
    
    results['无约束最优'] = {
        'index': max_yield_idx,
        'temperature': optimal_all['温度'],
        'conversion': optimal_all['乙醇转化率(%)'],
        'selectivity': optimal_all['C4烯烃选择性(%)'],
        'yield': optimal_all['C4烯烃收率'],
        'catalyst_id': optimal_all['催化剂组合编号'],
        'co_loading': optimal_all['Co_loading'],
        'ethanol_conc': optimal_all['ethanol_conc'],
        'total_catalyst': optimal_all['总催化剂质量'],
        'ratio': optimal_all['Co_SiO2_HAP_ratio_corrected']
    }
    
    # 约束优化
    if len(constrained_data) > 0:
        max_yield_idx_const = constrained_data['C4烯烃收率'].idxmax()
        optimal_const = constrained_data.loc[max_yield_idx_const]
        
        results['约束最优'] = {
            'index': max_yield_idx_const,
            'temperature': optimal_const['温度'],
            'conversion': optimal_const['乙醇转化率(%)'],
            'selectivity': optimal_const['C4烯烃选择性(%)'],
            'yield': optimal_const['C4烯烃收率'],
            'catalyst_id': optimal_const['催化剂组合编号'],
            'co_loading': optimal_const['Co_loading'],
            'ethanol_conc': optimal_const['ethanol_conc'],
            'total_catalyst': optimal_const['总催化剂质量'],
            'ratio': optimal_const['Co_SiO2_HAP_ratio_corrected']
        }
    
    # 输出结果
    for condition, result in results.items():
        print(f"\n{condition}:")
        print(f"  催化剂编号: {result['catalyst_id']}")
        print(f"  温度: {result['temperature']:.1f}°C")
        print(f"  乙醇转化率: {result['conversion']:.2f}%")
        print(f"  C4烯烃选择性: {result['selectivity']:.2f}%")
        print(f"  C4烯烃收率: {result['yield']:.2f}")
        print(f"  催化剂参数:")
        print(f"    Co负载量: {result['co_loading']:.1f}wt%")
        print(f"    乙醇浓度: {result['ethanol_conc']:.2f}ml/min")
        print(f"    总催化剂质量: {result['total_catalyst']:.0f}mg")
        print(f"    Co/SiO2-HAP比例: {result['ratio']:.2f}")
    
    return results

def analyze_top_performers(data, top_n=10):
    """分析top性能组合的特征"""
    
    print(f"\n分析前{top_n}个最优组合的特征:")
    print("="*50)
    
    # 无约束top组合
    top_all = data.nlargest(top_n, 'C4烯烃收率')
    
    # 约束top组合
    constrained_data = data[data['温度'] < 350]
    top_constrained = constrained_data.nlargest(top_n, 'C4烯烃收率') if len(constrained_data) > 0 else pd.DataFrame()
    
    analysis_results = {}
    
    for condition, top_data in [('无约束', top_all), ('约束', top_constrained)]:
        if len(top_data) == 0:
            continue
            
        stats = {
            'count': len(top_data),
            'yield_mean': top_data['C4烯烃收率'].mean(),
            'yield_std': top_data['C4烯烃收率'].std(),
            'temp_mean': top_data['温度'].mean(),
            'temp_std': top_data['温度'].std(),
            'conversion_mean': top_data['乙醇转化率(%)'].mean(),
            'selectivity_mean': top_data['C4烯烃选择性(%)'].mean(),
            'co_loading_mean': top_data['Co_loading'].mean(),
            'ethanol_conc_mean': top_data['ethanol_conc'].mean(),
            'total_catalyst_mean': top_data['总催化剂质量'].mean(),
            'ratio_mean': top_data['Co_SiO2_HAP_ratio_corrected'].mean()
        }
        
        analysis_results[condition] = stats
        
        print(f"\n{condition}条件前{top_n}组合特征:")
        print(f"  平均收率: {stats['yield_mean']:.2f} ± {stats['yield_std']:.2f}")
        print(f"  平均温度: {stats['temp_mean']:.1f} ± {stats['temp_std']:.1f}°C")
        print(f"  平均转化率: {stats['conversion_mean']:.2f}%")
        print(f"  平均选择性: {stats['selectivity_mean']:.2f}%")
        print(f"  平均Co负载量: {stats['co_loading_mean']:.2f}wt%")
        print(f"  平均乙醇浓度: {stats['ethanol_conc_mean']:.2f}ml/min")
        print(f"  平均催化剂质量: {stats['total_catalyst_mean']:.0f}mg")
        print(f"  平均Co/SiO2-HAP比例: {stats['ratio_mean']:.2f}")
    
    return analysis_results, top_all, top_constrained

def visualize_data_driven_results(data, optimal_results, top_all, top_constrained):
    """可视化数据驱动优化结果"""
    
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    
    # 收率分布对比
    ax1 = axes[0, 0]
    ax1.hist(data['C4烯烃收率'], bins=20, alpha=0.7, label='所有数据', color='lightblue')
    ax1.hist(data[data['温度'] < 350]['C4烯烃收率'], bins=20, alpha=0.7, label='温度<350°C', color='orange')
    ax1.axvline(optimal_results['无约束最优']['yield'], color='red', linestyle='--', label='无约束最优')
    if '约束最优' in optimal_results:
        ax1.axvline(optimal_results['约束最优']['yield'], color='green', linestyle='--', label='约束最优')
    ax1.set_xlabel('C4烯烃收率')
    ax1.set_ylabel('频数')
    ax1.set_title('C4烯烃收率分布')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # 温度vs收率散点图
    ax2 = axes[0, 1]
    ax2.scatter(data['温度'], data['C4烯烃收率'], alpha=0.6, s=30, label='所有数据')
    ax2.scatter(optimal_results['无约束最优']['temperature'], 
               optimal_results['无约束最优']['yield'], 
               color='red', s=100, marker='*', label='无约束最优')
    if '约束最优' in optimal_results:
        ax2.scatter(optimal_results['约束最优']['temperature'], 
                   optimal_results['约束最优']['yield'], 
                   color='green', s=100, marker='*', label='约束最优')
    ax2.axvline(350, color='orange', linestyle='--', alpha=0.7, label='温度约束')
    ax2.set_xlabel('温度(°C)')
    ax2.set_ylabel('C4烯烃收率')
    ax2.set_title('温度与收率关系')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 转化率vs选择性散点图
    ax3 = axes[0, 2]
    ax3.scatter(data['乙醇转化率(%)'], data['C4烯烃选择性(%)'], 
               c=data['C4烯烃收率'], cmap='viridis', alpha=0.6, s=30)
    ax3.scatter(optimal_results['无约束最优']['conversion'], 
               optimal_results['无约束最优']['selectivity'], 
               color='red', s=100, marker='*', label='无约束最优')
    if '约束最优' in optimal_results:
        ax3.scatter(optimal_results['约束最优']['conversion'], 
                   optimal_results['约束最优']['selectivity'], 
                   color='green', s=100, marker='*', label='约束最优')
    ax3.set_xlabel('乙醇转化率(%)')
    ax3.set_ylabel('C4烯烃选择性(%)')
    ax3.set_title('转化率与选择性关系')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # 创建colorbar
    cbar = plt.colorbar(ax3.collections[0], ax=ax3)
    cbar.set_label('C4烯烃收率')
    
    # Top10组合特征对比
    ax4 = axes[1, 0]
    if len(top_all) > 0:
        features = ['温度', 'Co_loading', 'ethanol_conc', '总催化剂质量']
        all_means = [top_all[f].mean() for f in features]
        const_means = [top_constrained[f].mean() for f in features] if len(top_constrained) > 0 else [0]*4
        
        x = np.arange(len(features))
        width = 0.35
        
        ax4.bar(x - width/2, all_means, width, label='无约束Top10', alpha=0.8)
        ax4.bar(x + width/2, const_means, width, label='约束Top10', alpha=0.8)
        ax4.set_xticks(x)
        ax4.set_xticklabels(['温度', 'Co负载量', '乙醇浓度', '催化剂质量'])
        ax4.set_ylabel('平均值')
        ax4.set_title('Top10组合特征对比')
        ax4.legend()
        ax4.grid(True, alpha=0.3)
    
    # 收率随温度变化趋势
    ax5 = axes[1, 1]
    temp_bins = pd.cut(data['温度'], bins=10)
    temp_yield_mean = data.groupby(temp_bins)['C4烯烃收率'].mean()
    temp_centers = [interval.mid for interval in temp_yield_mean.index]
    
    ax5.plot(temp_centers, temp_yield_mean.values, 'bo-', linewidth=2, markersize=6)
    ax5.axvline(350, color='orange', linestyle='--', alpha=0.7, label='温度约束')
    ax5.set_xlabel('温度(°C)')
    ax5.set_ylabel('平均C4烯烃收率')
    ax5.set_title('收率随温度变化趋势')
    ax5.legend()
    ax5.grid(True, alpha=0.3)
    
    # 催化剂参数重要性分析
    ax6 = axes[1, 2]
    # 计算各参数与收率的相关系数
    corr_vars = ['温度', 'Co_loading', 'ethanol_conc', '总催化剂质量', 'Co_SiO2_HAP_ratio_corrected']
    correlations = [data[var].corr(data['C4烯烃收率']) for var in corr_vars]
    
    colors = ['red' if c < 0 else 'blue' for c in correlations]
    bars = ax6.barh(range(len(corr_vars)), correlations, color=colors, alpha=0.7)
    ax6.set_yticks(range(len(corr_vars)))
    ax6.set_yticklabels(['温度', 'Co负载量', '乙醇浓度', '催化剂质量', 'Co/SiO2-HAP比例'])
    ax6.set_xlabel('与收率的相关系数')
    ax6.set_title('参数重要性分析')
    ax6.grid(True, alpha=0.3)
    ax6.axvline(0, color='black', linestyle='-', alpha=0.5)
    
    plt.tight_layout()
    save_plot("数据驱动优化结果")
    plt.show()

# 主分析流程
def main():
    """数据驱动优化主流程"""
    
    print("问题3 - 数据驱动优化分析")
    print("="*50)
    
    # 1. 数据准备
    data = prepare_optimization_data()
    
    # 2. 重建回归模型
    models = rebuild_regression_models(data)
    
    # 3. 找到最优组合
    optimal_results = find_optimal_combinations(data)
    
    # 4. 分析top性能组合
    analysis_results, top_all, top_constrained = analyze_top_performers(data)
    
    # 5. 可视化结果
    visualize_data_driven_results(data, optimal_results, top_all, top_constrained)
    
    # 6. 保存结果
    print("\n保存分析结果:")
    
    # 保存最优组合
    optimal_df = pd.DataFrame(optimal_results).T
    save_result(optimal_df, "数据驱动最优组合")
    
    # 保存top组合
    save_result(top_all, "无约束Top10组合")
    if len(top_constrained) > 0:
        save_result(top_constrained, "约束Top10组合")
    
    # 保存分析统计
    analysis_df = pd.DataFrame(analysis_results).T
    save_result(analysis_df, "Top组合统计分析")
    
    print(f"\n数据驱动优化分析完成")
    return optimal_results, models, data

if __name__ == "__main__":
    optimal_results, models, data = main()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, LassoCV, ElasticNet, ElasticNetCV
from sklearn.svm import SVR
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel, Matern
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, cross_val_score, RepeatedKFold, learning_curve
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

# 设置随机种子和输出格式
np.random.seed(42)
plt.rcParams['axes.unicode_minus'] = False

# 创建结果文件夹
import os
PROBLEM_NUM = 2
os.makedirs(f'results/问题{PROBLEM_NUM}', exist_ok=True)

def save_result(data, description, file_format='csv'):
    """保存重要结果到文件"""
    if file_format == 'csv':
        filename = f"results/问题{PROBLEM_NUM}/{description}.csv"
        data.to_csv(filename, encoding='utf-8-sig', index=False)
    elif file_format == 'excel':
        filename = f"results/问题{PROBLEM_NUM}/{description}.xlsx"
        data.to_excel(filename, index=False, engine='openpyxl')
    print(f"已保存: {filename}")
    return filename

def save_plot(description, dpi=300, format='png'):
    """保存图表到文件"""
    filename = f"results/问题{PROBLEM_NUM}/{description}.{format}"
    plt.savefig(filename, dpi=dpi, bbox_inches='tight')
    print(f"已保存: {filename}")
    return filename

def load_data(filename, encoding='utf-8-sig'):
    """标准化数据读取"""
    filepath = f"data/{filename}"
    if filename.endswith('.csv'):
        return pd.read_csv(filepath, encoding=encoding)
    elif filename.endswith('.xlsx'):
        return pd.read_excel(filepath, engine='openpyxl')
    else:
        raise ValueError(f"不支持的文件格式: {filename}")

def prepare_data():
    """数据准备和特征工程"""
    raw_data = load_data("附件1.csv")
    data = raw_data.copy()
    
    # 计算C4烯烃收率
    data['C4烯烃收率'] = data['乙醇转化率(%)'] * data['C4烯烃选择性(%)'] / 100
    
    # 创建HAP存在标志
    data['含HAP'] = data['HAP_mass'] > 0
    
    # 修正Co_SiO2_HAP_ratio计算
    data['Co_SiO2_HAP_ratio_corrected'] = np.where(
        data['HAP_mass'] > 0,
        data['Co_SiO2_mass'] / data['HAP_mass'],
        -1
    )
    
    print(f"数据准备完成: {len(data)}个样本, {len(data.columns)}个变量")
    return data

def add_physics_features(data):
    """添加基于物理原理的特征"""
    enhanced_data = data.copy()
    
    # 阿伦尼乌斯方程相关特征
    enhanced_data['温度_倒数'] = 1 / (enhanced_data['温度'] + 273.15)
    
    # 关键交互项
    enhanced_data['温度_催化剂质量'] = enhanced_data['温度'] * enhanced_data['总催化剂质量']
    enhanced_data['温度_Co负载量'] = enhanced_data['温度'] * enhanced_data['Co_loading']
    
    # 处理无穷大值
    enhanced_data = enhanced_data.replace([np.inf, -np.inf], np.nan)
    
    return enhanced_data

def create_model_ensemble():
    """创建适合小样本的模型集合"""
    models = {
        'ElasticNetCV': ElasticNetCV(cv=5, random_state=42, max_iter=2000),
        'RidgeCV': RidgeCV(alphas=[0.1, 1.0, 10.0, 100.0, 1000.0], cv=5),
        'LassoCV': LassoCV(cv=5, random_state=42, max_iter=2000),
        'SVR_RBF': SVR(kernel='rbf', C=100, gamma='scale'),
        'SVR_Poly': SVR(kernel='poly', degree=2, C=100, gamma='scale'),
        'GaussianProcess': GaussianProcessRegressor(
            kernel=ConstantKernel(1.0) * RBF(length_scale=1.0),
            alpha=1e-6,
            random_state=42
        ),
        'Poly2_Ridge': Pipeline([
            ('poly', PolynomialFeatures(degree=2, interaction_only=True)),
            ('ridge', Ridge(alpha=10.0))
        ]),
        'RandomForest_Small': RandomForestRegressor(
            n_estimators=50,
            max_depth=5,
            min_samples_split=10,
            min_samples_leaf=5,
            random_state=42
        )
    }
    
    return models

def enhanced_regression_analysis(data):
    """增强的多模型回归分析"""
    analysis_data = data[data['含HAP']].copy()
    
    # 处理多重共线性：移除高相关性变量
    predictor_vars = ['温度', 'Co_loading', 'ethanol_conc', '总催化剂质量', 'Co_SiO2_HAP_ratio_corrected']
    target_vars = ['乙醇转化率(%)', 'C4烯烃选择性(%)', 'C4烯烃收率']
    
    # 添加基于物理原理的特征
    analysis_data = add_physics_features(analysis_data)
    
    # 更新特征列表
    enhanced_features = predictor_vars + ['温度_倒数', '温度_催化剂质量', '温度_Co负载量']
    
    all_models = {}
    all_performance = []
    
    for target in target_vars:
        print(f"\n{target}多模型对比分析:")
        
        # 准备数据
        X = analysis_data[enhanced_features].select_dtypes(include=[np.number])
        y = analysis_data[target]
        
        # 处理缺失值
        valid_mask = ~(X.isnull().any(axis=1) | y.isnull())
        X_clean = X[valid_mask]
        y_clean = y[valid_mask]
        
        print(f"有效样本数: {len(X_clean)}")
        
        # 数据标准化
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X_clean)
        X_scaled_df = pd.DataFrame(X_scaled, columns=X_clean.columns)
        
        # 定义模型
        models = create_model_ensemble()
        
        # 严格的交叉验证
        cv = RepeatedKFold(n_splits=5, n_repeats=3, random_state=42)
        
        target_results = {}
        
        for name, model in models.items():
            print(f"训练 {name}...")
            
            # 交叉验证
            cv_scores = cross_val_score(model, X_scaled_df, y_clean, cv=cv, scoring='r2')
            cv_rmse = cross_val_score(model, X_scaled_df, y_clean, cv=cv, 
                                    scoring='neg_root_mean_squared_error')
            
            # 训练完整模型
            model.fit(X_scaled_df, y_clean)
            
            # 存储结果
            target_results[name] = {
                'model': model,
                'cv_r2_mean': cv_scores.mean(),
                'cv_r2_std': cv_scores.std(),
                'cv_rmse_mean': -cv_rmse.mean(),
                'cv_rmse_std': cv_rmse.std(),
                'cv_scores': cv_scores
            }
            
            print(f"  交叉验证 R²: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
            print(f"  交叉验证 RMSE: {-cv_rmse.mean():.4f} ± {cv_rmse.std():.4f}")
            
            # 添加到总体性能表
            all_performance.append({
                '目标变量': target,
                '模型': name,
                '交叉验证R²': cv_scores.mean(),
                'R²标准差': cv_scores.std(),
                '交叉验证RMSE': -cv_rmse.mean(),
                'RMSE标准差': cv_rmse.std()
            })
        
        # 找到最佳模型
        best_model_name = max(target_results.keys(), 
                            key=lambda x: target_results[x]['cv_r2_mean'])
        
        print(f"最佳模型: {best_model_name}")
        print(f"最佳性能: R² = {target_results[best_model_name]['cv_r2_mean']:.4f}")
        
        # 存储目标变量的所有模型
        all_models[target] = {
            'models': target_results,
            'best_model': best_model_name,
            'scaler': scaler,
            'features': enhanced_features,
            'X_clean': X_clean,
            'y_clean': y_clean
        }
    
    return all_models, pd.DataFrame(all_performance)

def analyze_model_performance(all_models, performance_df):
    """分析模型性能"""
    print("\n模型性能汇总:")
    
    # 按目标变量分组显示
    for target in ['乙醇转化率(%)', 'C4烯烃选择性(%)', 'C4烯烃收率']:
        target_data = performance_df[performance_df['目标变量'] == target]
        target_data_sorted = target_data.sort_values('交叉验证R²', ascending=False)
        
        print(f"\n{target}:")
        print(f"{'模型':<20} {'R²':<8} {'±':<8} {'RMSE':<8} {'±':<8}")
        print("-" * 50)
        
        for _, row in target_data_sorted.iterrows():
            print(f"{row['模型']:<20} {row['交叉验证R²']:<8.4f} {row['R²标准差']:<8.4f} "
                  f"{row['交叉验证RMSE']:<8.4f} {row['RMSE标准差']:<8.4f}")
    
    # 找到整体最佳模型
    best_models = {}
    for target in ['乙醇转化率(%)', 'C4烯烃选择性(%)', 'C4烯烃收率']:
        target_data = performance_df[performance_df['目标变量'] == target]
        best_row = target_data.loc[target_data['交叉验证R²'].idxmax()]
        best_models[target] = best_row['模型']
        
        print(f"\n{target}最佳模型: {best_row['模型']} (R² = {best_row['交叉验证R²']:.4f})")
    
    return best_models

def visualize_model_comparison(all_models, performance_df):
    """可视化模型对比结果"""
    
    # 创建子图
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    # 1. 模型性能对比条形图
    ax1 = axes[0, 0]
    
    # 为每个目标变量创建分组条形图
    targets = ['乙醇转化率(%)', 'C4烯烃选择性(%)', 'C4烯烃收率']
    models = performance_df['模型'].unique()
    
    x = np.arange(len(models))
    width = 0.25
    
    for i, target in enumerate(targets):
        target_data = performance_df[performance_df['目标变量'] == target]
        target_data = target_data.set_index('模型').reindex(models)
        
        ax1.bar(x + i*width, target_data['交叉验证R²'], width, 
               label=target.replace('(%)', ''), alpha=0.8)
    
    ax1.set_xlabel('模型')
    ax1.set_ylabel('交叉验证 R²')
    ax1.set_title('模型性能对比')
    ax1.set_xticks(x + width)
    ax1.set_xticklabels(models, rotation=45, ha='right')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # 2. 模型稳定性对比
    ax2 = axes[0, 1]
    
    for i, target in enumerate(targets):
        target_data = performance_df[performance_df['目标变量'] == target]
        target_data = target_data.set_index('模型').reindex(models)
        
        ax2.bar(x + i*width, target_data['R²标准差'], width, 
               label=target.replace('(%)', ''), alpha=0.8)
    
    ax2.set_xlabel('模型')
    ax2.set_ylabel('R² 标准差')
    ax2.set_title('模型稳定性对比')
    ax2.set_xticks(x + width)
    ax2.set_xticklabels(models, rotation=45, ha='right')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 3. 热力图显示所有模型的R²
    ax3 = axes[1, 0]
    
    pivot_data = performance_df.pivot(index='模型', columns='目标变量', values='交叉验证R²')
    
    sns.heatmap(pivot_data, annot=True, cmap='YlOrRd', ax=ax3, fmt='.3f')
    ax3.set_title('模型性能热力图')
    ax3.set_xlabel('目标变量')
    ax3.set_ylabel('模型')
    
    # 4. 箱线图显示交叉验证分布
    ax4 = axes[1, 1]
    
    # 选择收率模型的交叉验证分数分布
    yield_models = all_models['C4烯烃收率']['models']
    
    box_data = []
    labels = []
    
    for name, model_info in yield_models.items():
        box_data.append(model_info['cv_scores'])
        labels.append(name)
    
    box_plot = ax4.boxplot(box_data, labels=labels, patch_artist=True)
    
    # 设置颜色
    colors = plt.cm.Set3(np.linspace(0, 1, len(box_data)))
    for patch, color in zip(box_plot['boxes'], colors):
        patch.set_facecolor(color)
    
    ax4.set_ylabel('交叉验证 R²')
    ax4.set_title('C4烯烃收率模型交叉验证分布')
    ax4.tick_params(axis='x', rotation=45)
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    save_plot("多模型对比结果")
    plt.show()

def generate_model_insights(all_models, performance_df):
    """生成模型洞察报告"""
    insights = {}
    
    for target in ['乙醇转化率(%)', 'C4烯烃选择性(%)', 'C4烯烃收率']:
        target_data = performance_df[performance_df['目标变量'] == target]
        
        # 找到最佳模型
        best_model = target_data.loc[target_data['交叉验证R²'].idxmax()]
        
        # 找到最稳定模型
        most_stable = target_data.loc[target_data['R²标准差'].idxmin()]
        
        # 计算改进幅度
        if 'RidgeCV' in target_data['模型'].values:
            baseline_r2 = target_data[target_data['模型'] == 'RidgeCV']['交叉验证R²'].iloc[0]
            improvement = best_model['交叉验证R²'] - baseline_r2
        else:
            improvement = 0
        
        insights[target] = {
            'best_model': best_model['模型'],
            'best_r2': best_model['交叉验证R²'],
            'most_stable': most_stable['模型'],
            'stability': most_stable['R²标准差'],
            'improvement': improvement
        }
    
    # 打印洞察报告
    print("\n模型洞察报告:")
    
    for target, info in insights.items():
        print(f"\n{target}:")
        print(f"  最佳模型: {info['best_model']} (R² = {info['best_r2']:.4f})")
        print(f"  最稳定模型: {info['most_stable']} (标准差 = {info['stability']:.4f})")
        if info['improvement'] > 0:
            print(f"  相对改进: +{info['improvement']:.4f}")
    
    return insights

def save_model_results(all_models, performance_df, insights):
    """保存模型结果"""
    
    # 保存性能对比表
    save_result(performance_df, "多模型性能对比")
    
    # 保存洞察报告
    insights_df = pd.DataFrame(insights).T
    save_result(insights_df, "模型洞察报告")
    
    # 保存每个目标变量的最佳模型信息
    for target in ['乙醇转化率(%)', 'C4烯烃选择性(%)', 'C4烯烃收率']:
        target_clean = target.replace('(%)', '').replace(' ', '_')
        
        models_info = all_models[target]['models']
        best_model_name = all_models[target]['best_model']
        
        # 创建模型详情表
        model_details = []
        for name, info in models_info.items():
            model_details.append({
                '模型名称': name,
                '交叉验证R²': info['cv_r2_mean'],
                'R²标准差': info['cv_r2_std'],
                '交叉验证RMSE': info['cv_rmse_mean'],
                'RMSE标准差': info['cv_rmse_std'],
                '是否最佳': name == best_model_name
            })
        
        model_details_df = pd.DataFrame(model_details)
        save_result(model_details_df, f"{target_clean}_模型详情")
    
    print(f"\n模型结果已保存，共生成 {3 + len(all_models)} 个结果文件")

def create_yield_predictor(all_models, target='C4烯烃收率'):
    """创建收率预测函数用于问题3优化"""
    model_info = all_models[target]
    best_model_name = model_info['best_model']
    best_model = model_info['models'][best_model_name]['model']
    scaler = model_info['scaler']
    
    def predict_yield(params):
        """
        params: [温度, Co_loading, ethanol_conc, 总催化剂质量, Co_SiO2_HAP_ratio_corrected]
        """
        # 添加物理特征
        enhanced_params = list(params)
        enhanced_params.append(1 / (params[0] + 273.15))  # 温度_倒数
        enhanced_params.append(params[0] * params[3])      # 温度_催化剂质量
        enhanced_params.append(params[0] * params[1])      # 温度_Co负载量
        
        X = scaler.transform([enhanced_params])
        return best_model.predict(X)[0]
    
    return predict_yield

# 主分析流程
def main():
    """主分析流程"""
    
    # 1. 数据准备
    print("步骤1: 数据准备与预处理")
    data = prepare_data()
    
    # 2. 增强的多模型回归分析
    print("\n步骤2: 增强的多模型回归分析")
    all_models, performance_df = enhanced_regression_analysis(data)
    
    # 3. 模型性能分析
    print("\n步骤3: 模型性能分析")
    best_models = analyze_model_performance(all_models, performance_df)
    
    # 4. 可视化结果
    print("\n步骤4: 可视化模型对比结果")
    visualize_model_comparison(all_models, performance_df)
    
    # 5. 生成洞察报告
    print("\n步骤5: 生成模型洞察报告")
    insights = generate_model_insights(all_models, performance_df)
    
    # 6. 保存结果
    print("\n步骤6: 保存模型结果")
    save_model_results(all_models, performance_df, insights)
    
    # 7. 创建优化预测函数
    print("\n步骤7: 创建问题3优化预测函数")
    yield_predictor = create_yield_predictor(all_models)
    
    print("\n多模型分析完成！")
    print("可以使用 all_models 和 yield_predictor 进行问题3的优化")
    
    return all_models, performance_df, insights, yield_predictor

if __name__ == "__main__":
    all_models, performance_df, insights, yield_predictor = main()
```
